# RMSNorm block

$RMS(x) = \sqrt{\frac{1}{d}\sum_{i=1}^d{x_i^2} + \epsilon}$

$RMSNorm(x_i) = \frac{x_i}{RMS(x)} * \gamma_i$



```python
import torch

print(torch.cuda.is_available())
```

```python
import torch.nn as nn

torch.manual_seed(42)

batch_size, seq_len, embed_dim = 16, 512, 128
tensor_input = torch.randn(batch_size, seq_len, embed_dim)
print(f"{tensor_input.shape=}")
print(f"{tensor_input=}")

rms = nn.RMSNorm(embed_dim, eps=1e-5)
rms.weight.data = torch.randn(embed_dim)
with torch.no_grad():
    tensor_output = rms(tensor_input)
print(f"{tensor_output.shape=}")
print(f"{tensor_output=}")
```

```python
tensor_input
```

```python
for name, p in rms.named_parameters():
    print(f"{name=}, {p.shape=}")
print(rms.weight)
print(rms.eps)
```

```python
def rmsnorm(x, weight, eps):
    rms_value = (tensor_input.pow(2).mean(axis=-1, keepdim=True) + eps).sqrt()
    return (x / rms_value) * weight

tensor_input_copy = tensor_input.clone()
tensor_output_copy = rmsnorm(x=tensor_input_copy, weight=rms.weight, eps=rms.eps)
```

```python
torch.allclose(tensor_output_copy, tensor_output)
```

```python
id(tensor_output), id(tensor_output_copy)
```

```python
(tensor_output - tensor_output_copy).mean().item()
(tensor_output - tensor_output_copy).abs().max().item()
```

```python
import numpy as np


def save_tensor_to_bin(filename: str, tensor: torch.Tensor):
    """Save list of 1D float32 tensors to binary file."""
    with open(filename, 'wb') as f:
        # Ensure float32 and contiguous
        arr = tensor.detach().cpu().numpy().astype(np.float32).flatten()
        f.write(arr.tobytes())
        print(f"Written {len(arr)} floats to {filename}")
```

```python
from pathlib import Path


abspath = Path().resolve().parent / "data" / "rmsnorm"

# SAVE INPUT TENSOR
save_tensor_to_bin(
    filename=str(abspath / "tensor_input.bin"),
    tensor=tensor_input
)

# SAVE RMS WEIGHT TENSOR
save_tensor_to_bin(
    filename=str(abspath / "rms_weight_data.bin"),
    tensor=rms.weight.data
)

# SAVE OUTPUT TENSOR
save_tensor_to_bin(
    filename=str(abspath / "tensor_output.bin"),
    tensor=tensor_output
)
```
